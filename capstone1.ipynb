```python
##grouping the Sectors

def sector_groups(series):
    if (series =="Telecommunications") | (series =="Technology") :
           return "Telecom, tech"
    elif (series =="Wholesalers") | (series =="Retailing") :
            return "Wholesalers,retailing"
    elif (series =="Transportation") | (series =="Motor Vehicles & Parts"):
            return "Transportation,motor vehicles,parts"
    elif (series =="Materials") | (series =="Chemicals"):
            return "Materials,chemicals"
    elif (series =="Apparel") | (series =="Industrials"):
            return "Industrials,apparel"
    else: 
            return series
    

data['sector_groups'] = data['Sector'].apply(sector_groups)

data['sector_groups'].value_counts(sort=True)
```

```python
from sklearn.cluster import KMeans 
from scipy.spatial.distance import cdist
```

```python
# elbow mathod for choosing the K for k means 
distortions = []
K = range(1,12)
for k in K:
    kmeanModel = KMeans(n_clusters=k,random_state=0).fit(normalized_data)
    kmeanModel.fit(normalized_data)
    distortions.append(sum(np.min(cdist(normalized_data, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / normalized_data.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# silhouette_score for choosing the K for k means
from sklearn.metrics import silhouette_score
range_n_clusters=list(range(2,12))
for n_clusters in range_n_clusters:
    cluster_k_means = KMeans (n_clusters=n_clusters,random_state=0)
    preds = cluster_k_means.fit_predict(normalized_data)
    centers = cluster_k_means.cluster_centers_

    score = silhouette_score (normalized_data, preds, metric='euclidean')
    print ("For n_clusters = {}, silhouette score is {})".format(n_clusters, score))
```

```python
# k means clustering 
kmeans_4 = KMeans(n_clusters=4 ,random_state=26)

predictions_4 = kmeans_4.fit_predict(normalized_data)

centers = kmeans_4.cluster_centers_
silhouette_avg = silhouette_score(normalized_data, predictions_4)
```

```python
#DBSCAN
from sklearn import cluster
from sklearn.cluster import DBSCAN

# instance of DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=40).fit(normalized_data)
# use DBSCAN's fit_predict to return clustering labels for the data
clustering_labels_1 = dbscan.fit_predict(normalized_data)
n_clusters_ = len(set(clustering_labels_1)) - (1 if -1 in clustering_labels_1 else 0)


print('Estimated number of clusters: %d' % n_clusters_)
n_clusters_
```

```python
# grid search on parameters for DBSCAN 
epsilon=[0.1,0.125,0.15,0.2,0.21,0.215,0.22,0.23,0.24,0.25]
min_points=[10,20,30,40,45,50]

for epsilon1 in epsilon:
    for min_point in min_points:
        dbscan = DBSCAN(eps=epsilon1, min_samples=min_point).fit(normalized_data)
        clustering_labels_1 = dbscan.fit_predict(normalized_data)
        n_clusters_ = len(set(clustering_labels_1)) - (1 if -1 in clustering_labels_1 else 0)
        if (n_clusters_==1 | n_clusters_==0): 
            continue
        score = silhouette_score (normalized_data, clustering_labels_1, metric='euclidean')
        print ("for epsilon = {} min_points = {} the n_clusters is  = {}, and the silhouette score is {})".
               format(epsilon1,min_point,n_clusters_,score))
        print (epsilon1,min_point)
        
```
